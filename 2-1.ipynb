```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files

# 파일 업로드 선택창 호출
uploaded = files.upload()

# 이미지 불러오기 (업로드한 파일명 그대로 사용)
img = cv.imread('Fig0309(a)(washed_out_aerial_image).tif')

# R 채널 추출
r = img[:, :, 2]

# OTSU 알고리즘을 이용한 이진화
t, bin_img = cv.threshold(r, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
print('OTSU 알고리즘이 찾은 최적 임곗값 =', t)

# 파워 변환 (y = 3.0, 4.0, 5.0)
r_norm = r / 255.0
power3 = np.uint8(np.clip(np.power(r_norm, 3.0) * 255, 0, 255))
power4 = np.uint8(np.clip(np.power(r_norm, 4.0) * 255, 0, 255))
power5 = np.uint8(np.clip(np.power(r_norm, 5.0) * 255, 0, 255))

# 4장 출력
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Otsu Binarization")
plt.imshow(bin_img, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Power Transform y=5.0")
plt.imshow(power5, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Power Transform y=4.0")
plt.imshow(power4, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Power Transform y=3.0")
plt.imshow(power3, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
