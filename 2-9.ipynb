```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# 이미지 업로드 호출
uploaded = files.upload()

# 이미지 불러오기
img = cv.imread('Fig0940(a)(rice_image_with_intensity_gradient).tif', cv.IMREAD_UNCHANGED)

# 1. Otsu 이진화 (자동 임계값 계산)
otsu_thresh_val, otsu = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

# 2. 히스토그램 이진화 (임계값 100)
histogram_thresh_val = 100
_, manual = cv.threshold(img, histogram_thresh_val, 255, cv.THRESH_BINARY)

# 3. 결과
titles = ['Original Image', 'Otsu Thresholding', 'Histogram Thresholding']
images = [img, otsu, manual]

plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
