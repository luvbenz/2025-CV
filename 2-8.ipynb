```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

#  
uploaded = files.upload()

#  
img = cv.imread('Fig0938(a)(cygnusloop_Xray_original).tif', cv.IMREAD_UNCHANGED)
gray = img

# Morphological Smoothing ( 3x3)
se = cv.getStructuringElement(cv.MORPH_RECT, (3, 3))
closing = cv.morphologyEx(gray, cv.MORPH_CLOSE, se)
morph_smooth = cv.morphologyEx(closing, cv.MORPH_OPEN, se)

# Smoothing Filter (  3x3)
smooth = cv.blur(gray, (3, 3))

# 
# Morphological Smoothing    +    
# Smoothing Filter       
titles = ['Original Image', 'Morphological Smoothing', 'Smoothing Filter']
images = [gray, morph_smooth, smooth]

plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
